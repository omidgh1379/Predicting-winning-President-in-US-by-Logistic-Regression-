**import the dataset and library**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/president_county_candidate.csv')
data
```

# **Preprocessing**

 **handle cat**

```python
data.info()
```

```python
data1 = data.replace({'won' : True} , 1).replace({'won' : False} , 0)
data1
```

```python
print(np.unique(data1.party)) 
print()        #number of unique value of party
print(data1.party.value_counts())
```

```python
data1.candidate.value_counts()
```

```python
data2 = data1.drop(['state' , 'county' ] , axis = 1)
data2
```

```python
data3 = pd.get_dummies(data2)
data3
```

**duplicated**

```python
data3.duplicated().sum()

data4 = data3.drop_duplicates().reset_index().drop('index' , axis = 1)
data4
```

**missing value**

```python
data4.isna().sum()
```

**handle imbalance data imbalanced data**

```python
data4.won.value_counts()
```

```python
X = data4.drop(columns = 'won' , axis = 1)

y = data4.iloc[: , 1]
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

X_smote , y_smote = smote.fit_resample(X , y)
```

```python
y_smote.value_counts()
```

**feature scaling**

```python
from sklearn.preprocessing import MinMaxScaler

min_max = MinMaxScaler()

X_mm = min_max.fit_transform(X_smote)
```

```python
X_mm = pd.DataFrame(X_mm , columns = X_smote.columns)
X_mm
```

# splitting the data in train test 

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test  =  train_test_split(X_mm , y_smote ,test_size = 0.3 , random_state = 42)
```

**training the model using Logistic regression**

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import GridSearchCV

model = LogisticRegressionCV()

```

**finding the best hyper parameters**

```python
model.get_params()
```

```python
weight = [ 5 , 8,9,10, 12, 20 ]

grid = GridSearchCV(model , param_grid = {'Cs' : weight , 'penalty' : ['l2']} , cv  = 3 ,return_train_score=True , scoring = 'f1')
grid.fit(X_train , y_train)
grid.best_params_
```

```python


model = LogisticRegressionCV(Cs = 10)



model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred , y_test)
```

```python
from sklearn.metrics import f1_score

f1_score(y_pred , y_test)
```

**checking that we have overfitting or not**

```python
y_pred_train = model.predict(X_train)

accuracy_score(y_pred_train , y_train)
```
